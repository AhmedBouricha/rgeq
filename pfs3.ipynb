```python
import tkinter as tk
import random

user_pick = None
comp_pick = None
attempt_count = 0

def play():
    global user_pick, comp_pick, attempt_count
    comp_pick = random.choice(["Pierre", "Papier", "Ciseaux"])

    resultat_text.set(f"Joueur : {user_pick} Ordinateur : {comp_pick}")

    if user_pick == comp_pick:
        label.config(text="Égalité !")
    elif ((user_pick == "Pierre" and comp_pick == "Ciseaux") or
          (user_pick == "Papier" and comp_pick == "Pierre") or
          (user_pick == "Ciseaux" and comp_pick == "Papier")):
        label.config(text="Le joueur gagne !")
    else:
        label.config(text="L'ordinateur gagne !")

    attempt_count += 1
    if attempt_count == 3:
        determine_winner()

def determine_winner():
    if label.cget("text") == "Le joueur gagne !":
        result_label.config(text="Le joueur a remporté la partie !")
    elif label.cget("text") == "L'ordinateur gagne !":
        result_label.config(text="L'ordinateur a remporté la partie !")
    else:
        result_label.config(text="Partie terminée avec une égalité !")

def exit_game():
    root.destroy()

def reset():
    global user_pick, comp_pick, attempt_count
    label.config(text="Pierre, papier ou ciseaux ?")
    resultat_text.set("")
    user_pick = None
    comp_pick = None
    attempt_count = 0
    result_label.config(text="")

def choisir_pierre():
    global user_pick
    user_pick = "Pierre"
    play()

def choisir_papier():
    global user_pick
    user_pick = "Papier"
    play()

def choisir_ciseaux():
    global user_pick
    user_pick = "Ciseaux"
    play()

root = tk.Tk()
root.title("Pierre-Papier-Ciseaux")
root.geometry("500x250")
root.configure(bg="pink")

resultat_text = tk.StringVar()
resultat_entry = tk.Entry(root, textvariable=resultat_text, state='disabled', width=40)
resultat_entry.pack()
resultat_entry.pack_forget()

bouton_pierre = tk.Button(root, text="Pierre", command=choisir_pierre)
bouton_pierre.pack()
bouton_pierre.pack_forget()

bouton_papier = tk.Button(root, text="Papier", command=choisir_papier)
bouton_papier.pack()
bouton_papier.pack_forget()

bouton_ciseaux = tk.Button(root, text="Ciseaux", command=choisir_ciseaux)
bouton_ciseaux.pack()
bouton_ciseaux.pack_forget()

bouton_reinitialiser = tk.Button(root, text="Réinitialiser le jeu", command=reset)
bouton_reinitialiser.pack()
bouton_reinitialiser.pack_forget()

result_label = tk.Label(root, text="", fg='black', bg='pink')
result_label.pack()

bouton_quitter = tk.Button(root, text="Quitter", command=exit_game)
bouton_quitter.pack()
bouton_quitter.pack_forget()

def show_game():
    label.pack()
    resultat_entry.pack()
    bouton_pierre.pack()
    bouton_papier.pack()
    bouton_ciseaux.pack()
    bouton_reinitialiser.pack()
    bouton_quitter.pack()

def play_game():
    label.pack()
    show_game()
    bouton_play.pack_forget()

label = tk.Label(root, text="Pierre, papier ou ciseaux ?", fg='black', bg='pink')
label.pack()
label.pack_forget()

bouton_play = tk.Button(root, text="Play", command=play_game)
bouton_play.pack()

root.mainloop()
```
