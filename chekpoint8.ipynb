```python
import requests                               # importer les informations d'une url 
from bs4 import BeautifulSoup

def parse_html_content(url):                             # demander à pythom de retirer le contenu de la page wikipédia
    page = requests.get(url)
    return page.content

def extract_article_title(html_content):                          # demander à python de montrer le titre de la page wikipédia
    soup = BeautifulSoup(html_content, "html.parser")
    title = soup.title.string
    return title

def extract_article_text(html_content):                          # demander à python de extraire tous les composants de la page
    soup = BeautifulSoup(html_content, "html.parser")
    headers = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
    return headers

def collect_every_link(html_content):                        # demander à python de collecter tous les liens
    soup = BeautifulSoup(html_content, "html.parser")
    links = soup.find_all('a')
    return links

def all_the_previous_functions(url):                           # rassemler tous les fonctions dans une seule fonction
    html = parse_html_content(url)
    article_title = extract_article_title(html)
    article_headers = extract_article_text(html)
    article_links = collect_every_link(html)

    print("Title:", article_title)                  # afficher le titre de la page "Hannibal Barca — Wikipédia" 


    print("Headers:")                       #afficher tous les composants de la page 
    for header in article_headers:
        print(header)

    print("Links:")                           # afficher tous les liens de la page
    for link in article_links:
        print(link.get('href'))

all_the_previous_functions("https://fr.wikipedia.org/wiki/Relativit%C3%A9_restreinte")        # faire le rapport entre ce code et l'url
```
