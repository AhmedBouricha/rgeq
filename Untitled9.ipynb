```python
import tkinter as tk
import random

user_pick = None
comp_pick = None
resultat = None

def jouer():
    global user_pick, comp_pick, resultat
    comp_pick = random.choice(["Pierre", "Papier", "Ciseaux"])
    
    resultat.config(text=f"Joueur : {user_pick} Ordinateur : {comp_pick}")
    resultat.pack()
    
    if user_pick == comp_pick:
        label.config(text="Égalité !")
    elif ((user_pick == "Pierre" and comp_pick == "Ciseaux") or
          (user_pick == "Papier" and comp_pick == "Pierre") or
          (user_pick == "Ciseaux" and comp_pick == "Papier")):
        label.config(text="Le joueur gagne !")
    else:
        label.config(text="L'ordinateur gagne !")

def choisir_pierre():
    global user_pick
    user_pick = "Pierre"
    jouer()

def choisir_papier():
    global user_pick
    user_pick = "Papier"
    jouer()

def choisir_ciseaux():
    global user_pick
    user_pick = "Ciseaux"
    jouer()

def reinitialiser():
    global user_pick, comp_pick
    label.config(text="Pierre, papier ou ciseaux ?")
    resultat.pack_forget()
    user_pick = None
    comp_pick = None

window = tk.Tk()
window.title("Pierre-Papier-Ciseaux")
window.geometry("720x480")
window.configure(bg="pink")

label = tk.Label(window, text="Pierre, papier ou ciseaux ?", fg='black', bg='pink')
label.pack()

resultat = tk.Label(window, text="", fg='black', bg='pink')

bouton_pierre = tk.Button(window, text="Pierre", command=choisir_pierre)
bouton_pierre.pack()

bouton_papier = tk.Button(window, text="Papier", command=choisir_papier)
bouton_papier.pack()

bouton_ciseaux = tk.Button(window, text="Ciseaux", command=choisir_ciseaux)
bouton_ciseaux.pack()

bouton_reinitialiser = tk.Button(window, text="Réinitialiser le jeu", command=reinitialiser)
bouton_reinitialiser.pack()

window.mainloop()
```
