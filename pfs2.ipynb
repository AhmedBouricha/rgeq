```python
import tkinter as tk
import random

user_pick = None
comp_pick = None
resultat = None
attempt_count = 0

def play():
    global user_pick, comp_pick, resultat, attempt_count
    comp_pick = random.choice(["Pierre", "Papier", "Ciseaux"])

    resultat.config(text=f"Joueur : {user_pick} Ordinateur : {comp_pick}")
    resultat.pack()

    if user_pick == comp_pick:
        label.config(text="Égalité !")
    elif ((user_pick == "Pierre" and comp_pick == "Ciseaux") or
          (user_pick == "Papier" and comp_pick == "Pierre") or
          (user_pick == "Ciseaux" and comp_pick == "Papier")):
        label.config(text="Le joueur gagne !")
    else:
        label.config(text="L'ordinateur gagne !")

    attempt_count += 1
    if attempt_count == 3:
        determine_winner()

def determine_winner():
    if label.cget("text") == "Le joueur gagne !":
        result_label.config(text="Le joueur a remporté la partie !")
    elif label.cget("text") == "L'ordinateur gagne !":
        result_label.config(text="L'ordinateur a remporté la partie !")
    else:
        result_label.config(text="Partie terminée avec une égalité !")

def exit_game():
    window.destroy()

def choisir_pierre():
    global user_pick
    user_pick = "Pierre"
    play()

def choisir_papier():
    global user_pick
    user_pick = "Papier"
    play()

def choisir_ciseaux():
    global user_pick
    user_pick = "Ciseaux"
    play()

def reset():
    global user_pick, comp_pick, attempt_count
    label.config(text="Pierre, papier ou ciseaux ?")
    resultat.pack_forget()
    user_pick = None
    comp_pick = None
    attempt_count = 0
    result_label.config(text="")

def create_game_window():
    global window, label, resultat, bouton_pierre, bouton_papier, bouton_ciseaux, bouton_reinitialiser, result_label, bouton_quitter, bouton_play

    window = tk.Tk()
    window.title("Pierre-Papier-Ciseaux")
    window.geometry("720x480")
    window.configure(bg="pink")

    label = tk.Label(window, text="Pierre, papier ou ciseaux ?", fg='black', bg='pink')
    label.pack()

    resultat = tk.Label(window, text="", fg='black', bg='pink')

    bouton_pierre = tk.Button(window, text="Pierre", command=choisir_pierre)
    bouton_pierre.pack()
    bouton_pierre.pack_forget()

    bouton_papier = tk.Button(window, text="Papier", command=choisir_papier)
    bouton_papier.pack()
    bouton_papier.pack_forget()

    bouton_ciseaux = tk.Button(window, text="Ciseaux", command=choisir_ciseaux)
    bouton_ciseaux.pack()
    bouton_ciseaux.pack_forget()

    bouton_reinitialiser = tk.Button(window, text="Réinitialiser le jeu", command=reset)
    bouton_reinitialiser.pack()
    bouton_reinitialiser.pack_forget()

    result_label = tk.Label(window, text="", fg='black', bg='pink')

    bouton_quitter = tk.Button(window, text="Quitter", command=exit_game)
    bouton_quitter.pack()

    bouton_play = tk.Button(window, text="Play", command=show_game)
    bouton_play.pack()

def show_game():
    bouton_pierre.pack()
    bouton_papier.pack()
    bouton_ciseaux.pack()
    bouton_reinitialiser.pack()
    result_label.pack()
    bouton_play.pack_forget()
    resultat.pack()

create_game_window()
window.mainloop()
```
